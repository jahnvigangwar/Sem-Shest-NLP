```python
corpus = gutenberg

doc = nlp(" ".join(word for word in corpus.words("austen-emma.txt") if word.isalpha()))
adjective_nouns = Counter((token.text, token.head.text) for token in doc if token.pos_ == "ADJ" and token.head.pos_ == "NOUN" and not token.is_stop and not token.head.is_stop)
most_common = adjective_nouns.most_common(10)

for pair, count in most_common:
  print(pair[0])
  print(pair[1])
  print("count= ", count)
  print()
```

```python
#Q3 Write a program to find out the Jaccard similarity between any two corpora.
```

```python
import nltk
nltk.download('gutenberg')
```

```python
def jaccard_similarity(text1, text2):
    intersection = text1.intersection(text2)
    union = text1.union(text2)
    score = len(intersection)/ len(union)
    return score
```

```python
text1 = set(gutenberg.words(gutenberg.fileids()[0]))
text2 = set(gutenberg.words(gutenberg.fileids()[1]))

ans = jaccard_similarity(text1, text2)
ans
```

```python
text4 = " Ravi and Raju are the best friends from school days. They wanted to go for a world tour and visit famous cities like Paris, London, Dubai, Rome etc and also they called their another friend Mohan to take part of this world tour. They started their journey from Hyderabad and spent next 3 months travelling all the wonderful cities in the world and cherish a happy moments!"
```
